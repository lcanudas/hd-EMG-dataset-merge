## Opening only a part of the database

You can load only the rows you are interested. You can do that by using the keyword `filters` at the function `read_parquet`. Below we load only one of the databases, with number id 4. 


```python
data = pd.read_parquet('../data/processed/dado.parquet', engine='pyarrow',filters=[('database_id','==',4)])
data
```

Below we load only the data that measured the EMG of the forearm..

```python
data= pd.read_parquet('../data/processed/dado.parquet', engine='pyarrow',filters=[('segment','==','forearm')])
data
```

These data are from two datasets, from the following articles:

```python
data['database URL'].unique()
```

From this selected data, we will pick the data of one trial. The trial selected will be from the database withj id number 3, subject 1, third trial from experimental  session 2.

```python
data_selected = data.query('subject==1 and trial==3 and database_id==3 and session==2 and task=="point"')
data_selected['task']
```

```python
fs = data_selected['frequency_sample (Hz)'].unique()[0]
t = np.arange(len(data_selected))/fs
plt.plot(t, data_selected['10'])
```
