## Opening only a part of the database

You can load only the rows you are interested. You can do that by using the keyword `filters` at the function `read_parquet`. Below we load only one of the databases, with number id 4. 


```python
data = pd.read_parquet('../data/processed/dado.parquet', engine='pyarrow',filters=[('database_id','==',4)])
data
```

Below we load only the data that measured the EMG of the forearm from subject 1 on trial 1 of session 1

```python
data= pd.read_parquet('../data/processed/dado.parquet', engine='pyarrow',filters=[[('segment','==','forearm'), 
                                                                                  ('trial','==',1),
                                                                                  ('session','==',1),
                                                                                  ('subject','==',1),
                                                                                  ('task','==','point')]])
data
```

```python
fs = data['frequency_sample (Hz)'].unique()[0]
t = np.arange(len(data))/fs
plt.plot(t, data['10'])
```
